```python
import os
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.utils import shuffle
from skimage import exposure

# img =cv2.imread(r"D:\Downloads\b.jpg")
img = cv2.imread(r"D:\Downloads\GTSRB_Final_Test_Images\GTSRB\Final_Test\Images\00001.ppm")
print(f"img shape: {img.shape}")

def MinMaxNorm(image, a=-.5, b=.5) :
    Xmin, Xmax = np.min(image), np.max(image)
    return (image-Xmin)*(b-a)/(Xmax-Xmin)

def AbsNorm(image, a=-.5, b=0.5, col_min=0, col_max=255) : # chuyen ve khoang 
    return (image-col_min)*(b-a)/(col_max-col_min)

def to_BGR(img):
    img = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
    return img;
def to_hsv(img):
    img = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
    return img

def to_yuv(img):
    img = cv2.cvtColor(img,cv2.COLOR_RGB2YUV)
    y,u,v = [img[:,:,i] for i in range(3)]
    print(f"y shape: {y.shape}")
#     y=img[:,:,0]
    y=cv2.cvtColor(y,cv2.COLOR_GRAY2BGR)
#     u=img[:,:,1]
    u=cv2.cvtColor(u,cv2.COLOR_GRAY2BGR)
#     v=img[:,:,2]
    v=cv2.cvtColor(v,cv2.COLOR_GRAY2BGR)
    
    return y,u,v
def to_gray_cv2(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#     print(img)
    return img

def preprocess_dataset(X):
    #Convert to grayscale, e.g. single Y channel => x shape (:,:)
    # cong thuc : Y =0.299R+0.587G+0.114B, U = -0.147R-0.289G+0.436B, V=0.615R-0.515G-0.100B
    # YUV
    X = 0.299*X[ :, :, 0] + 0.587*X[:, :, 1] + 0.114*X[:, :, 2]
#     Scale features to be in [0, 1]
    X = (X / 255.).astype(np.float32)
    # exposure.equalize_adapthist must value of x is range(0,1)
    X = exposure.equalize_adapthist(X) 

    # Add a single grayscale channel, if you want to show image, you need reshape X to (img_size,img_size), 
    #show more in plot_preprocess_dataset function
    # shape is returned  (num_img, img_size,img_size, 1)
    X = X.reshape(X.shape + (1,))
    return X

def plot_preprocess_dataset(img):
    x = preprocess_dataset(img)
    x_height = x.shape[0]
    x_width = x.shape[1]
    x_channel = x.shape[2]
    x=x.reshape(x_height,x_width)
    plt.imshow(x, cmap='gray')
    plt.show()

def plot_histogram_contrast(img):
    img = cv2.cvtColor(img,cv2.COLOR_RGB2YUV)
    y,u,v = [img[:,:,i] for i in range(3)]
    u=u.reshape(-1)
    y=y.reshape(-1)
    fig,axes = plt.subplots(1,2)
    for ax in axes:
        ax.axis('off')
    axes[0].hist(u,256 ,range=(0, 256), fc='k', ec='k')
    axes[0].set_title("low contrast")
    
    axes[1].hist(y,256 ,range=(0, 256), fc='k', ec='k')
    axes[1].set_title("hight contrast")
    
    plt.show()




if np.array_equal(to_BGR(img),to_hsv(img)):
    print('equal')
fig, ax = plt.subplots(1,6,figsize=(40,10))
for axes in ax :
    axes.axis('off')
ax[0].imshow(to_BGR(img))
ax[0].set_title("original")

ax[1].imshow(to_hsv(img))
ax[1].set_title('hsv')

ax[3].imshow(to_gray_cv2(img))
ax[3].set_title('gray')

plt.show()

fig, ax = plt.subplots(1,3,figsize=(40,10))
for axes in ax :
    axes.axis('off')
y,u,v = to_yuv(img)
ax[0].imshow(y)
ax[0].set_title("y")

ax[1].imshow(u)
ax[1].set_title('u')

ax[2].imshow(v)
ax[2].set_title('v')
plt.show()



plot_preprocess_dataset(img)
plot_histogram_contrast(img)

img = cv2.imread(r"D:\Downloads\GTSRB_Final_Test_Images\GTSRB\Final_Test\Images\00001.ppm",0)
plt.imshow(img,cmap='gray')
plt.show()

img = cv2.imread(r"D:\Downloads\GTSRB_Final_Test_Images\GTSRB\Final_Test\Images\00001.ppm")
plt.imshow(img[:,:,0],cmap='gray')

plt.show()
```

```python
#test brightness image

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage import data, img_as_float
from skimage import exposure
import skimage as sk
import numpy as np
import cv2
img = mpimg.imread(r"D:\Downloads\a.jpg")
img_ori =cv2.imread(r"D:\Downloads\GTSRB_Final_Test_Images\GTSRB\Final_Test\Images\00001.ppm")

#using for khoa luan
def test_brightness(img_ori):
#     np.random.seed(2)
    ratios = np.random.uniform(low=0.,high=4.,size=10)
    print("ratios: ",ratios)
    imgs = []
    for ratio in ratios:
        img = exposure.adjust_gamma(img_ori,ratio)
        imgs.append(img)

    fig,ax = plt.subplots(3,3)

    ax[0][0].imshow(cv2.cvtColor(img_ori,cv2.COLOR_RGB2BGR))
    ax[0][0].axis('off')
    ax = ax.flatten()
    for i in range(1,9,1):
        ax[i].axis('off')
        ax[i].imshow(cv2.cvtColor(imgs[i],cv2.COLOR_RGB2BGR))
        ax[i].set_title(ratios[i].round(2))

    plt.show()

    fig,axes = plt.subplots(3,3)
    axes=axes.flatten()
    print(type(ax))
    
    for idx,ax in enumerate(axes):
        ax.axis('off')
        if(idx==0):
            img = preprocess_dataset(img_ori)
        else:
            img = preprocess_dataset(imgs[idx])
        #reshape image from (img_height, img_width, 1) to (img_height,img_width)
        img=img.reshape(img.shape[0],img.shape[1])
        ax.set_title(idx)
        ax.imshow(img,cmap='gray')
    plt.show()
test_brightness(img_ori)
```
