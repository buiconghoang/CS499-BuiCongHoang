```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import os
import numpy as np
```

```python
path = r"/home/thangdn/Downloads/hoang/28bienbao" 
path_save_base = r"/home/thangdn/Downloads/hoang/20bienbao_img"
```

```python
os.listdir(path)
```

```python
img_28_trafficsign = []
label_28_trafficsign = []
for folder in os.listdir(path):
    folder_path = os.path.join(path, folder)
    path_save = os.path.join(path_save_base, folder)
    append =  True
    if not os.path.exists(path_save):
        os.makedirs(path_save)
    for image in os.listdir(folder_path):
        if image.endswith('.JPG'):
            image_name = image.split('.')[0]
            new_file_img = os.path.join(path_save,image_name +".png")
            file_path = os.path.join(folder_path, image)
#             print(file_path)
            img = cv2.imread(file_path)
            cv2.imwrite(new_file_img, img)
            if append:
                img_28_trafficsign.append(img)
                label_28_trafficsign.append(folder)
            append = False
```

```python
print(len(img_28_trafficsign))
img_28_trafficsign[0].shape
print([label_28_trafficsign[0]])
```

```python
fig,axes = plt.subplots(nrows=5,ncols=6,figsize=(20,20),clear=True)
plt.subplots_adjust(hspace=0.5,wspace =0.5)
axes = np.array(axes)
#     print(f"fig: {fig}, ax: {axes.shape}")
for i,ax in enumerate(axes.reshape(-1)):
    if i >=28:
        break;
    image = img_28_trafficsign[i]
    image = image[:,:,::-1]
    ax.imshow(image) #,aspect='auto'
    label = label_28_trafficsign[i]
    ax.set_title(label)
    ax.set_xticklabels([])
    ax.set_yticklabels([])
```
