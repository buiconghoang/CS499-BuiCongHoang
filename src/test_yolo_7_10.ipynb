```python
import cv2
import os
import cv2
import numpy as np
import pandas as pd
import pickle


# video_path =r"C:\Users\Bui Huy\traffic_sign_project\dataVN\video_test\Raw_Videos\video\test3.mp4"
# video_path = r"test3.mp4"
# video_path = r"20171026_IMG_0626Trim2_Trim.mp4"
# video_path = r"/home/thangdn/Downloads/hoang/drive-download-20190119T051140Z-001/20171026_IMG_0626Trim2_Trim.mp4"
# video_path = r"/home/thangdn/Downloads/hoang/drive-download-20190119T051140Z-001/20171026_IMG_0626Trim3_Trim.mp4"
# video_path = r"/home/thangdn/Downloads/hoang/drive-download-20190119T051140Z-001/IMG_0600Trim1.mp4"
# video_path = r"/home/thangdn/Downloads/hoang/drive-download-20190119T051140Z-001/IMG_0623Trim2.mp4"
# video_path = r"/home/thangdn/Downloads/hoang/drive-download-20190119T051140Z-001/IMG_0623Trim3.mp4"
# video_path = r"/home/thangdn/Downloads/hoang/drive-download-20190119T051140Z-001/MVI_1889Trim1.mp4"
# video_path = r"/home/thangdn/Downloads/hoang/drive-download-20190119T051140Z-001/MVI_1893Trim1.mp4"
# video_save = r"result/"
# name_video = video_path.split('/')[-1].split('.')[0];
# name_video = video_path.split('\\')[-1].split('.')[0];
# video_save = r"/home/thangdn/Downloads/hoang/result.mp4"

# video_path =r"../video_test/20171026_IMG_0626Trim2_Trim.mp4"
# video_path =r"../video_test/20171026_IMG_0626Trim3_Trim.mp4"
# video_path =r"../video_test/IMG_0600Trim1.mp4"
# video_path =r"../video_test/IMG_0623Trim2.mp4"
# video_path =r"../video_test/IMG_0623Trim3.mp4"
# video_path =r"../video_test/MVI_1889Trim1.mp4"
# video_path =r"../video_test/MVI_1893Trim1.mp4"
video_path =r"../video_test/IMG_0623Trim3.mp4"


video_save = r"../video_test/result_video/"
video_save = r"../video_test/result_detect/"
cap = cv2.VideoCapture(video_path)
name_video = video_path.split('/')[-1].split('.')[0];
# name_video = video_path.split('\\')[-1].split('.')[0];


colors = [tuple(255 * np.random.rand(3)) for _ in range(10)]
cap = cv2.VideoCapture(video_path)
(ret, frame) = cap.read()
fshape = frame.shape
fheight = fshape[0]
fwidth = fshape[1]
out = cv2.VideoWriter(video_save+name_video+'.mp4',-1, 20.0, (fwidth,fheight))

while True:
    stime = time.time()
    ret, frame = cap.read()
    if ret:
        predicts = tfnet.return_predict(frame)
#         print(frame.shape)
        for color, predict in zip(colors, predicts):
            tl = (predict['topleft']['x'], predict['topleft']['y'])
            br = (predict['bottomright']['x'], predict['bottomright']['y'])
            prob = predict['confidence']
            print(predict['label'])
            if prob > 0.7:
                img_raw = frame[tl[1]:br[1], tl[0]:br[0],:]
    #             cv2.imwrite('')
                img = conver_input(img_raw)
                label =   predict['label']+"_"+str(np.round(prob,3)) +"_"+ labels[int(predict['label'])]
                save_image(name_video,label,img_raw)
                frame = cv2.rectangle(frame, tl, br, color, 2)
                frame = cv2.putText(frame, label, tl, cv2.FONT_HERSHEY_COMPLEX, 1, (0, 0, 255), 2)
        cv2.imshow('frame', frame)
        out.write(frame)
        print('FPS {:.1f}'.format(1 / (time.time() - stime)))
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```
